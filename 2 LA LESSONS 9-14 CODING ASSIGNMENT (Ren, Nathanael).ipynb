<div class="alert alert-block alert-success"><b>Problem #14:</b>
Given: 



<p>&nbsp;</p>
    
Use the Gram-Schmidt procedure to produce an orthogonal matrix from the given set of vectors ... i.e., find an orthogonal basis for span<b>($v_{1}$, $v_{2}$)</b>
</div>

## Solution to Problem #14

```python
v1 = np.array([1, 1, 0])
v2 = np.array([1, 1, 1])

M = np.vstack([v1, v2])

for i in range(M.shape[1]):
    for j in range(i):
        M[:, i] -= np.dot(M[:, j], M[:, i]) * M[:, j]
    M[:, i] = M[:, i] / np.linalg.norm(M[:, i])
```

<div class="alert alert-block alert-success"><b>Problem #15:</b>
Given: 



<p>&nbsp;</p>
    
Use the Gram-Schmidt procedure to produce an orthogonal matrix from the given set of vectors ... i.e., find an orthogonal basis for span<b>($v_{1}$, $v_{2}$, $v_{3}$)</b>
</div>

## Solution to Problem #15

```python
v1 = np.array([1, 2, -2])
v2 = np.array([1, 0, -4])
v3 = np.array([5, 2, 0])

M = np.column_stack([v1, v2, v3])

for i in range(M.shape[1]):
    for j in range(i):
        M[:, i] -= np.dot(M[:, j], M[:, i]) * M[:, j]
    M[:, i] = M[:, i] / np.linalg.norm(M[:, i])
    
M
```

## Solution to Problem #16

```python
A = np.array([[1, 1, 0], [1, 0, 1], [0, 1, 1]])

# 1.
A

# 2.
Q, R = np.linalg.qr(A, "complete")
Q

# 3.
R

# 4.
np.linalg.inv(R) @ Q.T
```
